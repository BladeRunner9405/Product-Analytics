```python
import pandas as pd
from datetime import datetime
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

df = pd.read_csv('log-info.csv')
df
```

```python
def normal_distribution(mean, var):
    mu = mean
    sigma = var ** (1 / 2)

    x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
    plt.plot(x, stats.norm.pdf(x, mu, sigma), c='k')

    plt.axvline(mu, c='g')
    plt.axvline(mu - sigma, c='y')
    plt.axvline(mu + sigma, c='y')
    plt.axvline(mu - 2*sigma, c='r')
    plt.axvline(mu + 2*sigma, c='r')

    plt.show()
    return round(mu - sigma), round(mu + sigma)
```

```python
normal_distribution(df['age'].mean(), df['age'].var())
```

```python
male = df[df['gender'] == 'М']
normal_distribution(male['age'].mean(), male['age'].var())
```

```python
female = df[df['gender'] == 'Ж']
normal_distribution(female['age'].mean(), female['age'].var())
```
