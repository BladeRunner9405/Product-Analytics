```python
import pandas as pd
import datetime
import numpy as np
```

```python
df = pd.read_csv('actions_sessions.csv')
```

```python
df
```

```python
def check_confirmation(action, value):
    if action == 'confirmation':
        return value
    return 0
```

```python
def check_cart(action, value):
    if action == 'cart':
        return value
    return 0
```

```python
def action_count(action, action_check):
    if action == action_check:
        return 1
    return 0
```

```python
df['search_count'] = df.apply(lambda x: action_count(x.action, 'search'), axis=1)
df['product_count'] = df.apply(lambda x: action_count(x.action, 'product'), axis=1)
df['cart_count'] = df.apply(lambda x: action_count(x.action, 'cart'), axis=1)
df['checkout_count'] = df.apply(lambda x: action_count(x.action, 'checkout'), axis=1)
df['confirmation_count'] = df.apply(lambda x: action_count(x.action, 'confirmation'), axis=1)
```

```python
df
```

```python
df['confirmation_value'] = df.apply(lambda x: check_confirmation(x.action, x.value), axis=1)
```

```python
df['cart_value'] = df.apply(lambda x: check_cart(x.action, x.value), axis=1)
```

```python
sessionId_info = df.groupby(['session_id']).agg(uid=('uid', np.min),
                                                age=('age', np.min),
                                                gender=('gender', np.min),
                                                city=('city', np.min),
                                                value=('confirmation_value', np.sum),
                                                dt=('dt', np.min),
                                                cart_value=('cart_value', np.sum), 
                                                search_count=('search_count', np.sum), 
                                                product_count=('product_count', np.sum), 
                                                cart_count=('cart_count', np.sum),
                                                checkout_count=('checkout_count', np.sum), 
                                                confirmation_count=('confirmation_count', np.sum))
```

```python
sessionId_info
```

```python
sessionId_info = sessionId_info.reset_index()
```

```python
sessionId_info = sessionId_info.sort_values('session_id')
```

```python
sessionId_info
```

```python
sessionId_info.to_csv('sessionId_info.csv', index=False)
```
